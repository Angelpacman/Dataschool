# When should i use "inplace" parameter in pandas?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('uforeports.csv')
ufo
```

```python
ufo.shape
```

```python
ufo.drop("City", axis = 1).head()
```

```python
ufo.head()
```

#### inplace
- Cuando hacemos un drop pero el resultado no es asignado a la variable lo que sea que hayamos quitado del df no es permanente, debido a que el parametro inplace por defecto es False. Cambiar este parametro a True haria que el df tome la nueva forma de lo que hayamos quitado con la funciÃ³n .drop()

```python
ufo.drop("City", axis = 1, inplace = True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any')
```

```python
ufo.shape
```

```python
ufo.head()
```

```python
ufo.dropna(how='any', inplace=True)
ufo.head()
```

- Otra forma de hacer permanente un df modificado sin usar inplace podria ser asignar la operacion al mismo df

```python
ufo.set_index('Time')
ufo.head()
```

```python
#De la forma anterior no modifica de manera permanente el df como vemos con este .head()
ufo.head()
```

- La manera de hacerlo permanente pero sin usar inplace serÃ­a:

```python
ufo = ufo.set_index('Time')
```

```python
#Hacemos otro head para confirmar que los cambios ya son permanentes en el df ufo
ufo.head()
```
